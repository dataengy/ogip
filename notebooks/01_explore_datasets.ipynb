# OGIP — exploring the ML-ready datasets

The pipeline (`make run`) produces Parquet under `.run/data/outputs/`. This notebook is the
primary DS-facing interface: load the datasets and eyeball the features.

- `games.parquet` — one row per resolved game (core dimension)
- `market_features.parquet` — FS-layer features (popularity, critic score, …)

```python
from pathlib import Path

import duckdb
import pandas as pd  # noqa: F401

OUT = next(p for p in (Path("../.run/data/outputs"), Path(".run/data/outputs")) if p.exists())
games = duckdb.sql(f"select * from read_parquet('{OUT}/games.parquet')").df()
games
```

```python
features = duckdb.sql(
    f"""select title, release_year, rating, metacritic, popularity_score, critic_score
        from read_parquet('{OUT}/market_features.parquet')
        order by popularity_score desc"""
).df()
features
```
